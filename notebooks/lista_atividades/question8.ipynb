```python
# Bibliotecas importadas:
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
# Carregando o dataset com os dados do arquivo csv.
boston_housing_dataset = pd.read_csv('BostonHousing.csv', encoding='utf-8')
```

```python
# Visualizando os primeiros dados da tabela:
boston_housing_dataset.head(10)
```

```python
# Realizando a limpeza de valores nulos:
boston_housing_dataset = boston_housing_dataset.dropna(subset=['rm'])
```

```python
# Separando a feature e a classe para o modelo:
feature, classe = boston_housing_dataset[['rm']], boston_housing_dataset['medv']
```

```python
# Separando os conjuntos de dados para treinamento e validação:
feature_treinamento, feature_validacao, classe_treinamento, classe_validacao = train_test_split(feature, classe, test_size=0.30)
```

```python
# Criando o modelo de Regressão Linear:
modelo_RL = LinearRegression()
modelo_RL.fit(feature, classe)
```

```python
# Fazendo a predição com os valores de validação:
classe_predicao = modelo_RL.predict(feature_validacao)
```

```python
# Métrica MAE:
mean_absolute_error(classe_validacao, classe_predicao)
```

```python
# Métrica R2:
r2_score(classe_validacao, classe_predicao)
```
