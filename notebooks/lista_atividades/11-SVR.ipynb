```python
# Bibliotecas necessárias:
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
# Carregando o dataset com os dados do arquivo csv.
boston_housing_dataset = pd.read_csv('../../data/BostonHousing.csv', encoding='utf-8')
```

```python
# Visualizando os primeiros dados da tabela:
boston_housing_dataset.head(10)
```

```python
# Realizando a limpeza de valores nulos:
boston_housing_dataset = boston_housing_dataset.dropna(subset=['rm'])
```

```python
# Separando as features e a classe para o modelo:
features, classe = boston_housing_dataset[['rm']], boston_housing_dataset['medv']
```

```python
# Separando os conjuntos de dados para treinamento e validação:
features_treinamento, features_validacao, classe_treinamento, classe_validacao = train_test_split(features, classe)
```

```python
# Treinando o modelo de regressão do Support Vector Machines:
modelo_SVR = SVR()
modelo_SVR.fit(features, classe)
```

```python
# Realizando a predição com os dados de validação:
classe_predicao = modelo_SVR.predict(features_validacao)
```

```python
# Métrica do Mean Absolute Error:
print(f"Média Absoluta de Erro: {mean_absolute_error(classe_validacao, classe_predicao):.2f}")
```

```python
# Métrica do Mean Squared Error:
print(f"Média Quadrática de Erro: {mean_squared_error(classe_validacao, classe_predicao):.2f}")
```
