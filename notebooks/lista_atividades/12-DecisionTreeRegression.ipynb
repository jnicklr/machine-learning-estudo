Utilize a métrica R² e MAE para avaliação do desempenho.

```python
# Bibliotecas necessárias:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
# Carregando o dataset com os dados do arquivo csv.
bike_sharing_dataset = pd.read_csv('../../data/london-bikes.csv', encoding='utf-8')
```

```python
# Visualizando os primeiros dados do dataset:
bike_sharing_dataset.head(10)
```

```python
# Removendo os dados que não são necessários:
bike_sharing_dataset = bike_sharing_dataset.drop(['timestamp'], axis=1)
```

```python
# Separando as features e a classe para o modelo:
features, classe = bike_sharing_dataset.drop(['cnt'], axis=1), bike_sharing_dataset['cnt']
```

```python
# Separando os conjuntos de dados de treinamento e validação:
features_treinamento, features_validacao, classe_treinamento, classe_validacao = train_test_split(features, classe, test_size=0.30)
```

```python
# Treinando o modelo de regressão com uma DecisionTree:
modelo_DTR = DecisionTreeRegressor()
modelo_DTR.fit(features, classe)
```

```python
# Realizando a predição com os dados de validação:
classe_predicao = modelo_DTR.predict(features_validacao)
```

```python
# Métrica R2:
print(f"R² Score: {r2_score(classe_validacao, classe_predicao):.2f}")
```

```python
# Métrica Mean Absolute Error:
print(f"Média Absoluta de Erro: {mean_absolute_error(classe_validacao, classe_predicao):.2f}")
```
