# K-Nearest Neighbours Regression

10) Aplique KNN para realizar uma tarefa de regressão em um conjunto de dados como
Airfoil Self-Noise. Compare o desempenho do modelo com a regressão linear usando MSE.

```python
# Bibliotecas necessárias:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Carregando o dataset com os dados do arquivo csv:
airfoil_dataset = pd.read_csv('../../data/AirfoilSelfNoise.csv', encoding='utf-8')
```

```python
# Visualizando os dados do dataset:
airfoil_dataset.head(10)
```

```python
# Separando as features e a classe para os modelos:
features, classe = airfoil_dataset.drop(['SSPL'], axis=1), airfoil_dataset['SSPL']
```

```python
# Separando os conjuntos de dados de treinamento e validação:
features_treinamento, features_validacao, classe_treinamento, classe_validacao = train_test_split(features, classe, test_size=0.30)
```

```python
# Criando os modelos de KNN e Regressão Linear:
modelo_KNN = KNeighborsRegressor()
modelo_KNN.fit(features_treinamento, classe_treinamento)
modelo_RL = LinearRegression()
modelo_RL.fit(features_treinamento, classe_treinamento)
```

```python
# Realizando a predição dos modelos:
classe_predicao_KNN = modelo_KNN.predict(features_validacao)
classe_predicao_RL = modelo_RL.predict(features_validacao)
```

```python
# Realização das métricas e comparação entre os modelos:
```

```python
print(f"Mean Squared Error (KNN): {mean_squared_error(classe_validacao, classe_predicao_KNN):.4f}")
print(f"Mean Squared Error (Linear Regression): {mean_squared_error(classe_validacao, classe_predicao_RL):.4f}")
```
