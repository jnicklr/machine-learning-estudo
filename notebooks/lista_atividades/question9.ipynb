```python
# Bibliotecas importadas:
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
# Carregando o dataset com os dados do arquivo csv gerado pelo GPT:
dataFrame = pd.read_csv('consumo_energia_sintetico.csv', encoding='utf-8')
```

```python
# Visualização dos dados:
dataFrame.head(10)
```

```python
# Separando as features e a classe para o modelo de regressão linear múltiplo:
features, classe = dataFrame.drop(['consumo_energia'], axis=1), dataFrame['consumo_energia']
```

```python
# Seprando os conjuntos de dados para treinamento e validação:
features_treinamento, features_validacao, classe_treinamento, classe_validacao = train_test_split(features, classe, test_size=0.30)
```

```python
# Treinando o modelo de Regressão Linear:
modelo_RL = LinearRegression()
modelo_RL.fit(features, classe)
```

```python
# Realizando a predição com os dados de validação:
classe_predicao = modelo_RL.predict(features_validacao)
```

```python
# Métrica do MSE:
mean_squared_error(classe_validacao, classe_predicao)
```

```python
# Métrica do R2:
r2_score(classe_validacao, classe_predicao)
```
