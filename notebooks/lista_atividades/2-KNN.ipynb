# K-Nearest Neighbours

2) Construa um modelo KNN para prever a classe de um conjunto de dados (por
exemplo, Wine). Utilize validação cruzada para ajustar o número de vizinhos e avalie o
desempenho com a métrica F1-Score.

```python
# Bibliotecas necessárias:
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# Carrega o dataset com os dados do arquivo csv.
wine_dataset = pd.read_csv("../../data/wine_dataset.csv", encoding="utf-8")
```

```python
# Visualizando os primeiros dados para verificar a situação.
wine_dataset.head(10)
```

```python
# Mapeando os dados do tipo de vinho para valores numéricos.
wine_dataset['style'] = wine_dataset['style'].replace({'white': 0, 'red': 1}).infer_objects()
```

```python
# Verificando o resultado após o tratamento.
wine_dataset.head(10)
```

```python
# Definindo as features e a classe do modelo:
features, classe = wine_dataset.drop(['style'], axis = 1), wine_dataset['style']
```

```python
# Separando os dados de treinamento e de validação:
x_treino, x_val, y_treino, y_val = train_test_split(features, classe, test_size=0.30)
```

```python
# Treinando o modelo de KNN:
modelo_KNN = KNeighborsClassifier()
modelo_KNN.fit(x_treino, y_treino)
```

```python
# Predição do conjunto de treinamento.
y_pred_treino = modelo_KNN.predict(x_val)
```

```python
# Utilização da métrica do F1 Score:
print(f"F1-Score do Modelo: {f1_score(y_val, y_pred_treino)}")
```

```python
# Realizando a validação cruzada dos dados:
scores = cross_val_score(modelo_KNN, features, classe, cv=5, scoring='accuracy')
```

```python
print("Acurácia da Validação Cruzada: %0.2f" % (scores.mean()))
```
